```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
delivery = pd.read_csv('https://raw.githubusercontent.com/tobbyGithub/ML-CSC715/master/Delivery_fleet.csv',sep='\t')
```

```python
delivery.head()
```

## Visualizing the Dataset

```python

plt.figure()
plt.title('Distance Vs Speeding')
plt.plot(delivery.Distance_Feature, delivery.Speeding_Feature,'ko', c='maroon')
plt.ylabel('Speeding Feature')
plt.xlabel('Distance Feature')
plt.ylim(0,100)
```

```python
plt.scatter(delivery.Distance_Feature, delivery.Speeding_Feature)
plt.ylabel('Speeding Feature')
plt.xlabel('Distance Feature')
plt.ylim(0,100)
```

```python
# Importing the K-Means Model
```

```python
from sklearn.cluster import KMeans
```

## KMeans with number of clusters set to 2

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(delivery)
```

```python
#plot graph of data using two of the columns in dataset
plt.figure()
f1 = delivery['Distance_Feature'].values
f2 = delivery['Speeding_Feature'].values
X = np.matrix(list(zip(f1,f2)))
kmeans = KMeans(n_clusters=2).fit(X) #create kmeans model with number of clusters = 2 and train the model

#plot graph of data after kmeans clustering
# plot centroid 1
h1,=plt.plot(f1[kmeans.labels_==0],f2[kmeans.labels_==0],'go')
plt.plot(np.mean(f1[kmeans.labels_==0]),np.mean(f2[kmeans.labels_==0]),'g*',markersize=20,mew=3)
print(np.mean(f1[kmeans.labels_==0]),np.mean(f2[kmeans.labels_==0]))

# plot centroid 2
h2,=plt.plot(f1[kmeans.labels_==1],f2[kmeans.labels_==1],'bo')
plt.plot(np.mean(f1[kmeans.labels_==1]),np.mean(f2[kmeans.labels_==1]),'b*',markersize=20,mew=3)
print(np.mean(f1[kmeans.labels_==1]),np.mean(f2[kmeans.labels_==1]))

plt.ylabel('Speeding Feature')
plt.xlabel('Distance Feature')
plt.legend([h1,h2],['Group 1','Group 2'], loc='upper left')
plt.show()

```

## KMeans with number of clusters set to 4

```python
plt.figure()
kmeans = KMeans(n_clusters=4).fit(X)

h1,=plt.plot(f1[kmeans.labels_==0],f2[kmeans.labels_==0],'go')
plt.plot(np.mean(f1[kmeans.labels_==0]),np.mean(f2[kmeans.labels_==0]),'g*',markersize=20,mew=3)

h2,=plt.plot(f1[kmeans.labels_==1],f2[kmeans.labels_==1],'bo')
plt.plot(np.mean(f1[kmeans.labels_==1]),np.mean(f2[kmeans.labels_==1]),'b*',markersize=20,mew=3)

h3,=plt.plot(f1[kmeans.labels_==2],f2[kmeans.labels_==2],'mo')
plt.plot(np.mean(f1[kmeans.labels_==2]),np.mean(f2[kmeans.labels_==2]),'m*',markersize=20,mew=3)

h4,=plt.plot(f1[kmeans.labels_==3],f2[kmeans.labels_==3],'ro')
plt.plot(np.mean(f1[kmeans.labels_==3]),np.mean(f2[kmeans.labels_==3]),'r*',markersize=20,mew=3)

plt.ylabel('Speeding Feature')
plt.xlabel('Distance Feature')
plt.legend([h1,h2,h3,h4],['Group 1','Group 2','Group 3','Group 4'], loc='upper left')
plt.show()
```
